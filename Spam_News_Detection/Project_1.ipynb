```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
ps = WordNetLemmatizer()
stopword = stopwords.words('english')
nltk.download('wordnet')
```

```python
def clear_row(row):
  row = row.lower()
  row = re.sub('[^a-zA-Z]',' ',row)
  token = row.split()
  news = [ps.lemmatize(word) for word in token if not word in stopword]
  cleaned_new = ' '.join(news)
  return cleaned_new
```

```python
data_set['text']
```

```python
data_set['text'] = data_set['text'].apply(lambda x : clear_row(x))
```

```python
data_set['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=50000,lowercase=False,ngram_range=(1,2))
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label = train_test_split(x,y,random_state = 20)
```

```python
vect_train_data = vectorizer.fit_transform(train_data)
```

```python
vect_train_data = vect_train_data.toarray()
```

```python
type(vect_train_data)
```

```python
vect_test_data = vectorizer.fit_transform(test_data)
```

```python
vect_test_data = vect_test_data.toarray()
```

```python
type(vect_test_data)
```

```python
training_data = pd.DataFrame(vect_train_data,columns=vectorizer.get_feature_names_out())
```

```python
testing_data = pd.DataFrame(vect_test_data,columns=vectorizer.get_feature_names_out())
```

```python
testing_data
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(training_data,train_label)
```

```python
y_pred = clf.predict(testing_data)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
accuracy_score(test_label,y_pred)
```
